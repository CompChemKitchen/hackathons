```python
def Calc_MQNs(mols,pca=None):
    mols_final=[mol for mol in mols if mol is not None]
    MQNs=[rdMolDescriptors.MQNs_(mol) for mol in mols_final]
    if pca is None:
        pca=PCA(n_components=3)
        MQN_3D = pca.fit_transform(MQNs)
    else:
        MQN_3D = pca.transform(MQNs)
    MQN_3D_df = pd.DataFrame(MQN_3D, columns=['PC1','PC2','PC3'])
    MQN_names=['MQN{0}'.format(x) for x in range(1,len(MQNs[0])+1)]
    MQNs_df=pd.DataFrame(MQNs, columns=MQN_names)
    data=pd.concat([MQN_3D_df, MQNs_df],axis=1)
    return data, pca,mols_final

def Plot_MQNs(data,colorcol):
    pp = sns.pairplot(data=data, 
                  hue=colorcol, # <== 😀 Look here!
                  height=1.8, aspect=1.8, 
                  vars=['PC1', 'PC2', 'PC3'],
                  #palette={"red": "#FF9999", "white": "#FFE888"},
                  plot_kws=dict(edgecolor="black", linewidth=0.5))
```
